# Custom nonlinearity profile - example of a Gaussian profile modulated by the absolute value of sine

```python
import numpy as np
import matplotlib.pyplot as plt
from path_finder_lib import find_path
from test_path_finder_lib import test_sine
```

### Function that returns a Gaussian nonlinearity profile modulated with a sine wave to have phase matching

```python
def gaus_sin(s,sigma=1,Lambda=0.1):
    return  np.exp(-s**2/(sigma**2)/2)*(np.sin(2*np.pi*s/Lambda))
```

### Function that returns a sinusoidal nonlinearity profile


```python
def d_sin(s,Lambda=0.1):
    return  np.sin(2*np.pi*s/Lambda)
```

### Example 1 - sine wave nonlinearity (k=1)

```python
x,z,deff,s=find_path(d_sin,50000,delta=0.00001,k=1);

plt.plot(z,x)
plt.axis('equal')
plt.xlabel('x (mm)')
plt.ylabel('z (mm)')
plt.title('Waveguide top-view')
#plt.savefig('my_plot.png', dpi=400)
plt.show()

plt.plot(s,deff)
plt.xlabel('s (mm)')
plt.ylabel('deff')
plt.title('Nonlinear profile')
```

### Example 2 - Gaussian nonlinearity modulated with a sine wave (k=2)

```python
x,z,deff,s=find_path(gaus_sin,300000,k=2,Pini=-3);

plt.plot(z,x)
#plt.axis('equal')
plt.xlabel('x (mm)')
plt.ylabel('z (mm)')
plt.title('Waveguide top-view')
#plt.savefig('my_plot.png', dpi=400)
plt.show()

plt.plot(s,deff)
plt.xlabel('s (mm)')
plt.ylabel('deff')
plt.title('Nonlinear profile')
```
